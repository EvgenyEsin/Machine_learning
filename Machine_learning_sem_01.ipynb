Вам дан набор данных с фактическими значениями целевой переменной и предсказанными значениями:


Фактические значения

actual_values = np.array([100, 150, 200, 250])


Предсказанные значения

predicted_values = np.array([90, 140, 180, 270])

Задача студентов состоит в расчете среднеквадратичной ошибки (MSE) и средней абсолютной ошибки (MAE) для данного набора данных без использования модели. Студенты должны использовать правильные значения фактической переменной и предсказанные значения для расчета функций потерь. Они также должны проанализировать результаты и объяснить, какие значения функции потерь указывают на лучшую точность предсказания.

```python
import numpy as np
```

```python
actual_values = np.array([100, 150, 200, 250])
predicted_values = np.array([90, 140, 180, 270])
```

```python
mse = np.mean((actual_values - predicted_values) ** 2)
mae = np.mean(abs(actual_values - predicted_values))

print('MSE = ', mse)
print('MAE = ', mae)
```

 Cредняя абсолютная ошибка (MAE), вычисляет среднее значение суммы абсолютных разностей между прогнозируемыми и фактическими значениями.


Преимущество

MAE более устойчив к выбросам по сравнению со среднеквадратичной ошибкой (MSE), поскольку учитывает абсолютную разницу, уменьшая влияние чрезвычайно больших ошибок.
Потери MAE легко интерпретировать, поскольку они представляют собой среднюю величину ошибок, что упрощает доведение производительности модели до заинтересованных сторон.

Недостаток

MAE одинаково обрабатывает все ошибки, независимо от их величины. Это может быть недостатком в случаях, когда важно различать небольшие и большие ошибки.
Градиент MAE является постоянной величиной, которая может замедлить сходимость во время оптимизации, особенно по сравнению с MSE, где градиент уменьшается по мере уменьшения ошибки.

Среднеквадратичная ошибка (MSE) похожа на потерю MAE, но вместо вычисления абсолютных разностей, как мы делали в MAE, вычисляет среднее значение квадратов разностей между прогнозируемыми и фактическими значениями.. Основная идея возведения в квадрат заключается в том, чтобы наказать модель за большую разницу, чтобы модель избегала больших различий.

Преимущества:

Подходит для задач регрессии, где конечной целью является минимизация квадратических различий между прогнозируемыми и фактическими значениями.
Дифференцируемые и выпуклые.

Недостатки:

Чувствительны к выбросам из-за операции возведения в квадрат, которая искажает результаты в процессе оптимизации.
Не идеально подходит для задач, где за ошибки следует наказывать по-другому.
